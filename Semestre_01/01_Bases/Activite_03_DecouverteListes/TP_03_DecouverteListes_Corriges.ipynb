## Création et manipulation de listes

Rappels :
* Une liste est une structure de données contenant une série de valeurs. Les valeurs sont séparées par des virgules et encadrées par des crochets. Par exemple <tt>L = [1, 2, 3]</tt>.
* Pour accéder au e élément d’une liste on procède ainsi <tt>L[i]</tt>. Attention l’indexation commence à <tt>0</tt>.
* Pour accéder à la taille de liste on utilise <tt>len[L]</tt>.
* Pour ajouter un élément à une liste, on utilise <tt>L.append(4)</tt>.

## Recherche dans une liste
Nous allons rechercher si un nombre est dans une liste.
Commençons par générer une liste de nombre alétoires. 

```python
import random as rd
def generer_alea(nb: int) -> list :
    """
    Génération d'une liste nb de nombres aléatoires compris entre 0 inclus et nb exclus.
    """
    res = []
    for i in range(nb):
        res.append(rd.randrange(0,nb))
    return res
```

```python
def recherche_first_index_nb_01(nb:int, L:list):
    n=len(L)
    index=-1
    for i in range(n):
        if nb==L[i] and index==-1:
            index=i
    return(index)
```
